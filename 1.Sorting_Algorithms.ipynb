# **Sorting Algoritms**


```python
import random

random_list = [random.randint(0,100000) for i in range(100)]
```

```python
print("Original list : \n",random_list,"\n\n\n")

def selection_sort(random_list):
    n = len(random_list)
    if n <= 1:
        return random_list
    
    for i in range(n-1):
        # Assume the current index is the index of the minimum value
        min_index = i
        
        # Find the index of the smallest element in the remaining unsorted part of the list
        for j in range(i+1, n):
            if random_list[j] < random_list[min_index]:
                min_index = j
        
        # Swap the found minimum element with the first element (at index i)
        random_list[i], random_list[min_index] = random_list[min_index], random_list[i]
        

        
        
selection_sort(random_list)
print("Sorted list:\n",random_list)
```
